```python
#multiple condition means more than one condition if use multiple condition then use 
#Logical operator
#there are 2 types of Logical operator
# 1. and : & (ampercend)
# 2. or : | (pipeline)
# & : if all condition are true
# | : if anyone condition are true
```

```python
# To display those records whose make=mazda and fuel type=diesel
df[(df["make"]=="mazda")&(df["fuel-type"]=="diesel")]  #df[(condition1) & (condition2)]
```

```python
#conditional filtering
#To display those records whose price of car > 30000
df[df["price"]>30000]
```

```python
#to display those records whose price of car>30000 and fuel-type=diesel
df[(df["price"]>30000) & (df["fuel-type"]=="diesel")]  # & : and
```

```python
# To sepeate numerical type(int,float) and object type data means categorical data means 
# string type from dataset df
df_num=df.select_dtypes(["int64","float64"])
df_cat=df.select_dtypes("object")
```

```python
#LabelEncoder class (inbuilt class of python) is defined in inbuilt package sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
#create object of LabelEncoder class 
#fit_transform() inbuilt function of LabelEncoder class(passing arugments/parameters)
import warnings
warnings.filterwarnings('ignore')
for col in df_cat:
    le=LabelEncoder()
    df_cat[col]=le.fit_transform(df_cat[col])
```

```python
#to join means concatenate 2 dataframe df_num and df_cat and hold in new dataset df_new
df_new=pd.concat([df_num,df_cat],axis=1)  #axis=1 means column wise join 2 dataframe
```

```python
#first create dictionary
d1={"rno":[110,120,130,140,150],"name":["anu","renu","seema","deepa","jyoti"],"age":[23,21,24,21,25]}
```

```python
#Second dictionary create
d2={"rno":[110,130,140],"python":[67,89,70],"R":[60,70,80],"ML":[56,78,60]}
```

```python
#to convert dictionary to DataFrame of pandas
#First convert d1 dictionary to dataframe
student=pd.DataFrame(d1)
student
```

```python
#convert d2 dictionary to dataframe
marks=pd.DataFrame(d2)
marks
```

```python
merge_df=pd.merge(student,marks,on="rno")  
merge_df
```

```python
merge_df=pd.merge(student,marks,on="rno",how="outer")  #Nan means null value
merge_df
```
